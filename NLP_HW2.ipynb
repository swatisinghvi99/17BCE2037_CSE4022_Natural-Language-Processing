# Step 1. Import NLTK

```python
import nltk
nltk.download('all')
```

# Step 2. Import word_tokenize, sent_tokenize and pos_tag

```python
from nltk.tokenize import word_tokenize,sent_tokenize
from nltk import pos_tag
```

# Step 3. Read file

```python
file = open("gettysburg_addr.txt","r")
file = file.read()
file
```

# Step 4. Sentence Tokenization

```python
txt = sent_tokenize(file)
txt
```

# Step 5. Word Tokenization

```python
for i in range(len(txt)):
  txt[i] = word_tokenize(txt[i])

print(txt)
```

# Step 6. POS Tagging

```python
for i in range(len(txt)):
  txt[i] = pos_tag(txt[i])
print(txt)
```

# Step 7. Define grammar for Verb Phrases

```python
grammar = "VP: {<MD>?<VB.*><NP|PP>}"
```

# Step 8. Verb Phrases

```python
result = []
cp = nltk.RegexpParser(grammar)
for sentence in txt:
  ans = cp.parse(sentence)
  print(ans,'\n\n')
  result.append(ans)
```
