# Frequency Distribution and Conditional Frequency Distribution (of Presidential Inaugural Addresses)

# Step 1. Import NLTK

```python
import nltk
nltk.download('inaugural')
```

# Step 2. Import inaugural

```python
from nltk.corpus import inaugural
print(inaugural.fileids(),'\n\n')
```

```python
f_name= '1789-Washington.txt'
print(inaugural.words(fileids=f_name)[:10])
```

# Step 3. Text conversion

```python
txt = ' '.join(inaugural.words(fileids=f_name))
txt
```

# Step 4. Frequency Distribution

```python
freq_dist=nltk.FreqDist(inaugural.words(fileids=f_name))
freq_dist
```

# Step 5. Conditional Frequency Distribution

```python
from nltk.probability import ConditionalFreqDist
cfd = ConditionalFreqDist((len(word),word) for word in inaugural.words(fileids=f_name))
cfd[5]
```
