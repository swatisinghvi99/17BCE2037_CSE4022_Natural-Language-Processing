# Step 2: Simple Text Classification

```python
def gf(word):
  return {'last_letter':word[-1]}

gf('Obama')
```

```python
from nltk.corpus import names

names.words()
```

```python
len(names.words())
```

```python
labels = ([(name,'male') for name in names.words('male.txt')] + [(name, 'female')  for name in names.words('female.txt')])
labels
```

```python
import random
random.shuffle(labels)
labels
```

```python
features = [(gf(n),gender) for (n,gender) in labels]
train_set,test_set = features[5000:],features[:2000]
classifier = nltk.NaiveBayesClassifier.train(train_set)
```

```python
classifier.classify(gf('David'))
```

```python
classifier.classify(gf('Rocky'))
```

```python
classifier.classify(gf('Michael'))
```

```python
nltk.classify.accuracy(classifier,test_set)
```

# Step 3. Count Vectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer
vector = CountVectorizer(binary=True)
corpus = ["Tessaract is good optical character recognition engine  ", "optical character recognition is significant "]
vector.fit(corpus)
```

```python
vocab=vector.vocabulary_
vocab
```

```python
for key in sorted(vocab.keys()):
    print("{}:{}".format(key, vocab[key]))
```

```python
print(vector.transform(["This is a good optical illusion"]).toarray())
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector.transform(["Google Cloud Vision is a character recognition engine"]).toarray(), vector.transform(["OCR is an optical character recognition engine"]).toarray())
print(similarity)
```
