# Step 1. Import NLTK

```python
import nltk
nltk.download('all')
```

# Step 2. Basic Text Processing Pipeline

```python
text= '''Natural language processing (NLP) is a subfield of linguistics, 
computer science, information engineering, and artificial intelligence 
concerned with the interactions between computers and human (natural) languages, 
in particular how to program computers to process and analyze large amounts 
of natural language data. '''

for txt in text:
  sentences=nltk.sent_tokenize(text)
  print(sentences)
  for sentence in sentences:
    words = nltk.word_tokenize(sentence)
    tagged=nltk.pos_tag(words)
    print(tagged)
```

# Step 3. Import TweetTokenizer
